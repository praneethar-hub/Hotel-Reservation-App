```python
class hotelfourseasons:

    def __init__(self, sub_total='', room_rent=0, game_bill=0, restaurent_bill=0, laundary_bill=0, additional_charges=1800,name='',address='',check_in_date='',check_out_date='',room_no=101):

        print ("\n\n*****WELCOME TO Fourseasons Hotel*****\n")

        self.sub_total=sub_total

        self.restaurent_bill=restaurent_bill

        self.laundary_bill=laundary_bill

        self.game_bill=game_bill

        self.room_rent=room_rent
        self.additional_charges=additional_charges
        self.name=name
        self.address=address
        self.check_in_date=check_in_date
        self.check_out_date=check_out_date
        self.room_no=room_no
        
    def inputdata(self):
        self.name=input("\nEnter your name:")
        self.address=input("\nEnter your address:")
        self.check_in_date=input("\nEnter your check in date:")
        self.check_out_date=input("\nEnter your checkout date:")
        print("Your room no.:",self.room_no,"\n")
        
    def roomrent(self):

        print ("We have the following rooms for you:-")

        print ("1.  type A---->rs 6000 PN\-")

        print ("2.  type B---->rs 5000 PN\-")

        print ("3.  type C---->rs 4000 PN\-")

        print ("4.  type D---->rs 3000 PN\-")

        x=int(input("Enter Your Choice Please->"))

        n=int(input("For How Many Nights Did You Stay:"))

        if(x==1):

            print ("you have opted room type A")

            self.room_rent=6000*n

        elif (x==2):

            print ("you have opted room type B")

            self.room_rent=5000*n

        elif (x==3):

            print ("you have opted room type C")

            self.room_rent=4000*n

        elif (x==4):
            print ("you have opted room type D")

            self.room_rent=3000*n

        else:

            print ("please choose a room")

        print ("your room rent is =",self.room_rent,"\n")   
        
    def restaurentbill(self):

        print("*****RESTAURANT MENU*****")

        print("1.water----->Rs20",
              "2.tea----->Rs10",
              "3.breakfast combo--->Rs90",
              "4.lunch---->Rs110",
              "5.dinner--->Rs150",
              "6.Exit")


        while (1):

            choice=int(input("Enter your choice:"))


            if (choice==1):
                d=int(input("Enter the quantity:")) #2
                self.restaurent_bill=self.restaurent_bill+20*d

            elif (choice==2):
                d=int(input("Enter the quantity:"))
                self.restaurent_bill=self.restaurent_bill+10*d

            elif (choice==3):
                d=int(input("Enter the quantity:"))
                self.restaurent_bill=self.restaurent_bill+90*d

            elif (choice==4):
                d=int(input("Enter the quantity:"))
                self.restaurent_bill=self.restaurent_bill+110*d

            elif (choice==5):
                d=int(input("Enter the quantity:"))
                self.restaurent_bill=self.restaurent_bill+150*d

            elif (choice==6):
                break;
            else:
                print("Invalid option")

        print ("Total food Cost=Rs",self.restaurent_bill,"\n")
        
    def laundrybill(self):
        print ("******LAUNDRY MENU*******")

        print ("1.Shorts----->Rs3",
               "2.Trousers----->Rs4",
               "3.Shirt--->Rs5",
               "4.Jeans---->Rs6",
               "5.Girlsuit--->Rs8",
               "6.Exit")

        while (1):

            e=int(input("Enter your choice:"))

            if (e==1):
                f=int(input("Enter the quantity:"))
                self.laundary_bill=self.laundary_bill+3*f

            elif (e==2):
                f=int(input("Enter the quantity:"))
                self.laundary_bill=self.laundary_bill+4*f

            elif (e==3):
                f=int(input("Enter the quantity:"))
                self.laundary_bill=self.laundary_bill+5*f

            elif (e==4):
                f=int(input("Enter the quantity:"))
                self.laundary_bill=self.laundary_bill+6*f

            elif (e==5):
                f=int(input("Enter the quantity:"))
                self.laundary_bill=self.laundary_bill+8*f
            elif (e==6):
                break;
            else:

                print ("Invalid option")


        print ("Total Laundary Cost=Rs",self.laundary_bill,"\n")    
        
        
    def gamebill(self):
        print ("******GAME MENU*******")

        print ("1.Table tennis----->Rs60",
               "2.Bowling----->Rs80",
               "3.Snooker--->Rs70",
               "4.Video games---->Rs90",
               "5.Pool--->Rs50",
               "6.Exit")
        while (1):
            
            g=int(input("Enter your choice:"))
            
            
            if (g==1):
                h=int(input("No. of hours:"))
                self.game_bill=self.game_bill+60*h

            elif (g==2):
                h=int(input("No. of hours:"))
                self.game_bill=self.game_bill+80*h

            elif (g==3):
                h=int(input("No. of hours:"))
                self.game_bill=self.game_bill+70*h

            elif (g==4):
                h=int(input("No. of hours:"))
                self.game_bill=self.game_bill+90*h

            elif (g==5):
                h=int(input("No. of hours:"))
                self.game_bill=self.game_bill+50*h
            elif (g==6):
                break;

            else:

                print ("Invalid option")



        print ("Total Game Bill=Rs",self.game_bill,"\n") 
        
        
        
    def display(self):
        print ("******HOTEL BILL******")
        print ("Customer details:")
        print ("Customer name:",self.name)
        print ("Customer address:",self.address)
        print ("Check in date:",self.check_in_date)
        print ("Check out date",self.check_out_date)
        print ("Room no.",self.room_no)
        print ("Your Room rent is:",self.room_rent)
        print ("Your Food bill is:",self.restaurent_bill)
        print ("Your laundary bill is:",self.laundary_bill)
        print ("Your Game bill is:",self.game_bill)

        self.sub_total=self.room_rent+self.laundary_bill+self.game_bill+self.restaurent_bill

        print ("Your sub total bill is:",self.sub_total)
        print ("Additional Service Charges is",self.additional_charges)
        print ("Your grandtotal bill is:",self.sub_total+self.additional_charges,"\n")
        self.room_no+=1        
```
